
## *‚ö°!!!!Trata UF*

````python
#SUBISTITUIR PELO ESTADO DE MAIOR FREQUENCIA - BA
# Definindo a lista de estados do Nordeste brasileiro
nordeste = ['AL', 'BA', 'CE', 'MA', 'PB', 'PE', 'PI', 'RN', 'SE']

# Tratando o dom√≠nio categ√≥rico, substituindo pela moda (maior frequ√™ncia - 'BA')
group = dataset.groupby(['uf']).size()

# Ordenando os resultados em ordem descendente
result = group.sort_values(ascending=False)

# Pegando a moda (primeiro valor do resultado ordenado)
moda = result.index[0]

# Substituindo os valores n√£o pertencentes ao Nordeste pela moda ('BA')
dataset.loc[~dataset['uf'].isin(nordeste), 'uf'] = moda
````

## *‚ö°!!!!Trata MENSALIDADE*

````python
# Calculando a m√©dia das mensalidades condicionada √†s idades
mean_mensalidade = dataset.groupby('idade')['mensalidade'].mean()
````

````python
#visualiza media de salarios entre 45 e 50
mean_mean_mensalidade_filtrado = mean_mensalidade[(mean_mensalidade.index >= 45) & (mean_mensalidade.index <= 48)]
mean_mean_mensalidade_filtrado
````

````python
# Preenchendo as mensalidades nulas com as m√©dias correspondentes
for idade, media in mean_mensalidade.items():
    dataset.loc[(dataset['idade'] == idade) & (dataset['mensalidade'].isnull()), 'mensalidade'] = media
````

````python
#tipo_dados_mensalidade = muda para inteiro
dataset['mensalidade'] = dataset['mensalidade'].astype('int64')
````

##üè∏*!!!!Explora COPARTICIPACAO ANTES*



````python
#Participa√ß√£o - analise
dataset['participacao'].describe()
````

````python
srn.boxplot(dataset['participacao']).set_title('participacao')
````

````python
srn.histplot(dataset['participacao'],kde=True).set_title('participacao')
````

````python
#conta NANs
dataset['participacao'].isnull().sum()
````

````python
# Verificando quais valores na coluna 'participacao' est√£o nulos
valores_nulos_part = dataset['participacao'].isnull()
# Exibindo as linhas onde 'participacao' √© nulo
linhas_com_nulos_part = dataset[valores_nulos_part]
linhas_com_nulos_part
````

## *‚ö°!!!!Trata COPARTICIPACAO*

````python
#calular a mediana
mediana = sts.median(dataset['participacao'].dropna())
mediana
````

````python
#substituir NAN por mediana
dataset['participacao'] = dataset['participacao'].fillna(mediana)
````

````python
#calculo desvio padr√£o para participa√ß√£o - para verificar outiliers
desv = sts.stdev(dataset['participacao'])
desv
````

````python
#checamos os outliers existentes -perfis com participacao maior que 4 desvios padr√£o
linhas_part_outliers= dataset.loc[dataset['participacao'] > 4*desv ]
linhas_part_outliers
````

````python
# Definir outliers como 4 desvios padr√£o e atribuir a mediana
dataset['participacao'] = dataset['participacao'].apply(lambda x: mediana if x >= 4 * desv else x)
````

##‚úÖ*!!!!Explora COPARTICIPACAO P√ìS*

````python
#conta NANs
dataset['participacao'].isnull().sum()
````

````python
#visualiza substituri√ß√£o de participa√ß√£o nula
linhas_id_part = dataset[dataset['id'].isin(linhas_com_nulos_part['id'])]
linhas_id_part
````

````python
#checamos se h√° outliers
dataset.loc[dataset['participacao'] >= 4 * desv ]
````

````python
#Visualiza substitui√ß√£o pelo id
linhas_id_part = dataset[dataset['id'].isin(linhas_part_outliers['id'])]
linhas_id_part
````

````python
#tipo_dados_participacao = muda para inteiro
dataset['participacao'] = dataset['participacao'].astype('int64')
````
